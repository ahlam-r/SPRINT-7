Calculadora de l'índex de massa corporal

    Escriu una funció que calculi l'IMC ingressat per l'usuari/ària, és a dir, qui ho executi haurà d'ingressar aquestes dades. Pots obtenir més informació del seu càlcul en:

    -> Índice de masa corporal IMC que es y como se calcula.
    La funció ha de classificar el resultat en les seves respectives categories: 
        Bajo peso: menos de 18,5.
        Peso normal: 18,5-24,9.
        Sobrepeso: 25-29,9.
        Obesidad: más de 30.


```python
def calcular_imc(): #creo una funcion para calcular imc

    #Calculo IMC
    imc = peso / (altura ** 2) 

    # Clasificar el IMC
    if imc < 18.5:
        print("Tienes bajo peso.")
    elif 18.5 <= imc <= 24.9:
        print("Tienes un peso normal.")
    elif 25 <= imc <= 29.9:
        print("Tienes sobrepeso.")
    else:
        print("Tienes obesidad.")

# Solicitar datos al usuario
peso = float(input("Cual es tu peso en Kg: ")) 
altura = float(input("Cual es tu estatura en metros (ej 1.65): "))

# Muestro les dades introduides i el resultat del imc
print(f"\nTu peso es {peso}kg y altura {altura}m y tienes un IMC de {imc:.2f}")  

#executem la funcio
calcular_imc()
```

Convertidor de temperatures.

Existeixen diverses unitats de temperatura utilitzades en diferents contextos i regions. Les més comunes són Celsius (°C), Fahrenheit (°F) i Kelvin (K). També existeixen altres unitats com Rankine (°Ra) i Réaumur (°Re). Selecciona almenys 2 conversors, de tal manera que en introduir una temperatura retorni, com a mínim, dues conversions.Kelvin = 0°C + 273.15 = 273.15K

```python
def convertir_temperatura(temperatura, tipo):
    """Funcion que convierte la temperatura dada en las otras dos unidades y devuelve un mensaje."""

    #Si la temperatura esta en celsisus la convertira a Farenheit y Kelvin
    if tipo == "C":
        fahrenheit = temperatura * 9/5 + 32 #Formula de conversion Celsius a Farenheit
        kelvin = temperatura + 273.15 #Formula de conversion Celsius a Kelvin
        return f"Has elegido Celsisus, {temperatura}°C equivale a:\n{fahrenheit:.2f}°F (Fahrenheit)\n{kelvin:.2f}K (Kelvin)" 
        # este es el resultado que saldra por pantalla, y redondeamos 2 decimales
    
    #Si la temperatura esta en Farenheit la convertira a Celsius y Kelvin
    elif tipo == "F":
        celsius = (temperatura - 32) * 5/9 
        kelvin = (temperatura - 32) * 5/9 + 273.15
        return f"Has elegido Fahrenheit, {temperatura}°F equivale a:\n{celsius:.2f}°C (Celsius) \n{kelvin:.2f}K (Kelvin)"
    
     #Si la temperatura esta en Kelvin la convertira a Celsius y farenheit
    elif tipo == "K":
        celsius = temperatura - 273.15
        fahrenheit = (temperatura - 273.15) * 9/5 + 32
        return f"Has elegido Kelvin {temperatura} K equivale a:\n{celsius:.2f}°C (Celsius) \n{fahrenheit:.2f}°F (Fahrenheit)"
    
    
    # Si el usuario ingresa un tipo de temperatura no válido, mostramos un mensaje de error
    else:
        return "Error: Tipo de temperatura no válido."
    
#Se pide al usuario que eliga la temperatura a introducir y la convertimos en mayuscula para 
tipo = input("""Selecciona el tipo de temperatura que quieres convertir: C = Celsius, F = Fahrenheit, K = Kelvin""").upper()

# Pedir el valor de la temperatura
temperatura = float(input("Introduce el valor de la temperatura a convertir: "))

# Llamar a la función y mostrar el resultado
print("\n" + convertir_temperatura(temperatura, tipo))
```

- Exercici 3

Comptador de paraules d'un text.

Escriu una funció que donat un text, mostri les vegades que apareix cada paraula.


```python
def contador_de_palabras(text):
    """Analiza un texto y muestra la frecuencia de cada palabra de manera ordenada."""
    """Texto para testear: Tú me quieres alba,
    Ni un rayo de luna
    filtrado me haya.
    Ni una margarita
    se diga mi hermana.
    Tú me quieres nívea,
    tú me quieres blanca,
    tú me quieres alba."""

    # convertimos el texto a minúsculas para que detecte las mismas palabras
    text = text.lower()

    # eliminamos signos de puntuaciónpara evitar que afecten el conteo
    signos = ".,;:!?()[]{}\"'’“”" # Lista de signos que queremos eliminar
    for signo in signos:
        text = text.replace(signo, "")

    # dividimos el texto en palabras separadas usando split()
    palabras = text.split()

    # creamos un diccionario vacio para contar la frecuencia de cada palabra
    frec = {}
    # recorremos cada palabra y la agregamos al diccionario
    for palabra in palabras: #con get() contamos su frecuencia, si la palabra ya esta en el diccionario, la sumamos si no esta se inicia con 1
        frec[palabra] = frec.get(palabra, 0) + 1 


    # imprimimos un conteo general
    print(f"Total de palabras: {len(palabras)}") #len() nos devuelve lacCantidad total de palabras en el texto
    print(f"Palabras únicas: {len(frec)}\n") #Numero de palabras únicas en el texto

    # Imprimimos los resultados
    print("Frecuencia de cada palabra:\n")

    #recorremos el dicicionario de frec y mostramos cuantas veces aparace cada palabra
    for palabra, contador in frec.items():
        print(f"{palabra} : {contador}")  #imprime la palabra y su frecuencia
    
# Solicitamos el texto al usuario
texto_entrada = input("Introduce un texto: ")
contador_de_palabras(texto_entrada)
```

- Exercici 4

Diccionari invers.

Resulta que el client té una enquesta molt antiga que s'emmagatzema en un diccionari i els resultats els necessita al revés, és a dir, intercanviats les claus i els valors. Els valors i claus en el diccionari original són únics; si aquest no és el cas, la funció hauria d'imprimir un missatge d'advertiment.

```python
def invertir_dicc(dicc):
    "Funcion que invierte un diccionario asegurandose de que los valores sean unicos"

    #Comprueba valores duplicados con set
    if len(dicc.values()) != len(set(dicc.values())): #si el numero de valores del diccionario es diferente a el numero de valores conjunto set() sin duplicados
        print("Error: multiple keys for one value") #si el valor es diferente saldra en pantalla Error
        return None  # Retorno en caso de error

    #si el valores son iguales crea un diccionario invertido
    dicc_invertit = {value: key for key, value in dicc.items()}
    return dicc_invertit

# Ejemplo correcto
ex_1 = invertir_dicc({'a': 1, 'b': 2, 'c': 3})  
print(f"Resultado del diccionario invertido:\n",ex_1)  # Output: {1: 'a', 2: 'b', 3: 'c'}

# Ejemplo incorrecto (valores duplicados)
ex_2 = invertir_dicc({'x': 'apple', 'y': 'banana', 'z': 'banana'})  # Output: Error: multiple keys for one value
print(f"Resultado del diccionario invertido:\n",ex_2)  # Output: None

#ejemplo extra para testear
ex_3 = invertir_dicc({'perro': 'ladrido', 'gato': 'maullido', 'vaca': 'mugido'})
print(f"Resultado del diccionario invertido:\n", ex_3)  # Output: {'ladrido': 'perro', 'maullido': 'gato', 'mugido': 'vaca'}
```

Diccionari invers amb duplicats

Continuant amb l'exercici 4 del nivell 1: al client es va oblidar de comentar un detall i resulta que els valors en el diccionari original poden duplicar-se i més, per la qual cosa les claus intercanviades poden tenir duplicats. En aquest cas, en l'exercici anterior imprimies un missatge d'advertiment, ara, els valors del diccionari resultant hauran d'emmagatzemar-se com una llista. Tingues en compte que si és un valor únic no ha de ser una llista.



```python

def reverse_dictionary_plus(d):
    """Defino la funcion que invierte el diccionario, intercambia claves por valores.
        si un mismo valor tiene mutiples claves las almacena en una lista"""
    reversed_dict = {}  # Diccionario vacío para almacenar el resultado

    #el bucle con d.items() recorre cada clave-valor en el diccionario,  y intercambiar la clave con el valor.
    for key, value in d.items():
        if value in reversed_dict: # si el valor ya existe como clave en el reverse_dict
            if isinstance(reversed_dict[value], list): # comprueba si reversed_dict[value] ya es una lista
                reversed_dict[value].append(key)  #Si es una lista, significa que ya había varias claves asociadas, por lo que añadimos la nueva clave
            else:
                # Si value solo tenia una clave asociada, convertimos su valor en una lista para almacenar múltiples claves
                reversed_dict[value] = [reversed_dict[value], key]
        else:
            #Si el valor no está en reversed_dict, lo añadimos con su clave asociada
            reversed_dict[value] = key

    return reversed_dict #nos devuelve el diccionario invertido

# diccionarios usados para invertir
data_1 = {'x': 'apple', 'y': 'banana', 'z': 'banana'}
print('Caso 1: valores duplicados')
print(reverse_dictionary_plus(data_1))  # Resultado esperado{'apple': 'x', 'banana': ['y', 'z']}


data_2 = {'a': 1, 'b': 2, 'c': 3}
print('Caso 2: valores únicos')
print(reverse_dictionary_plus(data_2))  # Resultado esperado {1: 'a', 2: 'b', 3: 'c'}


```

Exercici 2

Conversió de tipus de dades

El client rep una llista de dades i necessita generar dues llistes, la primera on estaran tots els elements que es van poder convertir en flotants i l'altra on estan els elements que no es van poder convertir. Exemple de la llista que rep el client: [ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]

```python
#funcion que procesa la lista
def flatten_list(complex_list): 
  
    simple_list = []
    for element in complex_list:
        # Si el elemento es una cadena, número, o flotante, se agrega directamente
        if isinstance(element, (str, int, float)): 
            simple_list.append(element)
        # Si el elemento es un contenedor (tuple o lista), se procesa recursivamente
        else:
            simple_list.extend(flatten_list(element))
    return simple_list


def convert_to_float_and_flatten(data):
   
    # Aplanar la lista primero(todos los elementos son simples)
    flattened_data = flatten_list(data)
    
    # Listas para elementos convertidos y no convertidos
    converted = []
    error = []
    
    for item in flattened_data:
        try:
            # Intentar convertir a flotante
            converted.append(float(item))
        except ValueError:
            # Si no se puede convertir, agregar a la lista de errores
            error.append(item)
    
    return converted, error

# Ejemplo de uso
data = ['1.3', 'one', '1e10', 'seven', '3-1/2', 
        ('2', 1, 1.4, 'not-a-number'), 
        [1, 2, '3', '3.4']]

# Ejecutar la función
converted, error = convert_to_float_and_flatten(data)

# Imprimir los resultados
print('Converted elements:', converted)
print('Errors:', error)
```

Comptador i endreçador de paraules d'un text.

El client va quedar content amb el comptador de paraules, però ara vol llegir arxius TXT i que calculi la freqüència de cada paraula ordenades dins de les entrades habituals del diccionari segons la lletra amb la qual comencen, és a dir, les claus han d'anar de la A a la Z i dins de la A hem d'anar de la A la Z. Per exemple, per a l'arxiu "tu_me_quieres_blanca.txt" la sortida esperada seria:

```python
def comptador_de_paraules(text):
    """Analiza un texto y muestra la frecuencia de cada palabra de manera ordenada."""
    
    # Convertimos el texto a minusculas para que las palabras iguales sean contadas
    text = text.lower()

    # Eliminamos signos de puntuación
    signos = ".,;:!?()[]{}\"'’“”"  # Lista de signos que queremos eliminar
    for signo in signos:
        text = text.replace(signo, "")  # Reemplazamos los signos por un string vacío

    
    paraules = text.split() #Dividimos el texto en palabras
    paraules.sort()  # Ordenamos alfabéticamente las palabras

    # Creamos un diccionario para contar la frecuencia de cada palabra
    frec = {}
    for paraula in paraules:
        # Usamos el método get() para contar la frecuencia de las palabras
        frec[paraula] = frec.get(paraula, 0) + 1

    #Mostramos el resultado ordenado alfabéticamente
    for paraula, comptador in sorted(frec.items()):
        print(f"{paraula}: {comptador}")  #Imprimimos cada palabra con su frecuencia

#Abrimos el archivo donde se encuentra el texto
arxiu = open("tu_me_quieres_blanca.txt", "r")
text = arxiu.read()  #Leemos el archivo
arxiu.close()

comptador_de_paraules(text)  #Ejecutamos la función
```
